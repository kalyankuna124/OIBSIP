### Loading Dataset

```python
df = pd.read_csv('/content/drive/MyDrive/data science/Iris.csv')
```

```python
# Id column is not needed as it is not informative for the data training model 

df.drop('Id', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Species'].unique()
```

```python
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']
```

```python
x
```

```python
y
```

```python
y
```

### Train-Test Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

### Model Building, Training and Evaluation

```python
# Training the model with Logistic Regression and checking the score/accuracy

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
# Training the model with K Nearest Neighbour Clssification and checking the score/accuracy

from sklearn.neighbors import KNeighborsClassifier

model_KNN = KNeighborsClassifier(n_neighbors=10)

model_KNN.fit(X_train, y_train)

model_KNN.score(X_test, y_test)
```
